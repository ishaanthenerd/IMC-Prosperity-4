```python
PRODUCT_START = "PANEL_"
PRODUCT_SUBLIST = []
for p in default_products:
    if len(p) >= len(PRODUCT_START) and p[0:len(PRODUCT_START)] == PRODUCT_START:
        PRODUCT_SUBLIST.append(p)
LOW = 2e6
HIGH = 5e6

for PRODUCT in PRODUCT_SUBLIST:
    df_copy = df[(LOW <= df["timestamp"]) & (df["timestamp"] <= HIGH)].copy()
    df_copy = df_copy[df_copy["product"] == PRODUCT]

    df_copy["best_diff"] = df_copy["best_ask"] - df_copy["best_bid"]
    # plot_against_timestamp(df_copy, ["best_diff"], f"{PRODUCT}'s Best Diff")

    df_copy["rolling_std"] = df_copy["mid_price"].rolling(window = 1000).std()
    plot_against_timestamp(df_copy, ["rolling_std"], f"{PRODUCT}'s Rolling Std")
    plot_against_timestamp(df_copy, ["mid_price"], f"{PRODUCT}'s Mid Price")
    plt.show()
    print(PRODUCT, df_copy["mid_price"].mean(), df_copy["mid_price"].std())
```
