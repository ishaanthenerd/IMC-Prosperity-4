```python
N = 51

def normalize(x):
    return x * 5 + 670
    # if x <= 40:
    #     return x + 160
    # else:
    #     return (x - 40) + 250

arr = [normalize(i) for i in range(N)]

def calculate(arr, a, b):
    total = 0
    cnt = len(arr)
    for x in arr:
        if x <= a:
            total += a
        elif x <= b:
            total += b
        else:
            cnt -= 1
    return 920 * cnt - total

best = float("-inf")
best_nums = (0, 0)

def try_parameters(i, j):
    global best, best_nums
    result = calculate(arr, i, j)
    if result > best:
        best = result
        best_nums = (i, j)


for a in range(0, N - 1):
    i = normalize(a)
    for b in range(a + 1, N):
        j = normalize(b)
        try_parameters(i, j)

print(best)
print(best_nums)
print("Average:", best / N)
```
