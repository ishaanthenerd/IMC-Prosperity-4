```python
import numpy as np

def cdf(bid: int):
    return (1 + (bid - 670) // 5) / (1 + (920 - 670) // 5)
def under(bid: int):
    return cdf(bid) - (1/51 if bid % 5 == 0 else 0)
```

```python
def ev(bid_1: int, bid_2: int):
    under_1 = under(bid_1)
    under_2 = under(bid_2)
    return under_1 * (920 - bid_1) + (under_2 - under_1) * (920 - bid_2)
```

```python
max_profit = 0
best_bid_1 = 670
best_bid_2 = 670

for bid_1 in range(670, 921):
    for bid_2 in range(670, 921):
        profit = ev(bid_1, bid_2)
        if profit > max_profit:
            max_profit = profit
            best_bid_1 = bid_1
            best_bid_2 = bid_2
print("best bids of", best_bid_1, ",", best_bid_2, "--> profit of", round(max_profit, 2))
```
