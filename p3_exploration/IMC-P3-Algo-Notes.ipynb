**Algo Strategies from IMC Prosperity 3:**

**Round 1 - Market-Making** (chrispyroberts)

**Resin**
- Information
    - Fair value is 10000
- Strategy
    - Buy low, sell high
    - Use any orders exactly at 10000 to balance position

**Kelp**
- Information
    - Small drift, small volatility
- Strategy
    - Define the mid price as the average of the worst bid & ask, as the best bid & ask was often much more volatile
    - Widen our bid / ask to adjust for the volatile best bid / ask on the market (e.g. if our ask is 5100 and the best ask right now is 5115, change to 5115 $\implies$ adapt to info already there!)
    - Don't trade in a direction if both (1) our position already is in that direction and (2) we offer the best corresponding bid / ask $\implies$ risk aversion measure

**Squid ink**
- Information
    - Extremely volatile (100 shell swings in one step are normal)
    - Attempted strategies: Rolling z-scores, volatility breakouts
- Strategy
    - Cut down % of total position allocation (cuts PnL but does prevent massive risk)
    - Use rolling window on std; if std > threshold, enter a position fully opposite of the recent price move
    - Use a gridsearch on optimal window and threshold on anything that has this structure

**Round 2 - Arbitrage** (chrispyroberts)

**Croissants, Jams, Djembes, Baskets 1 & 2**
- Information
    - Basket 1: 6 Croissants, 3 Jams, 1 Djembe
    - Basket 2: 4 Croissants, 2 James
- Strategy
    - Compute premiums created by longing one basket and shorting it's components
    - Position limits were causing an issue $\implies$ experiment with different premium computations (in this case, what worked was using the difference in premiums)
    - Premimums were mean-reverting $\implies$ use mean reversion with hard-coded means and a window for standard deviations
    - Remaining position not being used $\implies$ either arbitrage more or go back to earlier market-making strategies

**Round 3 - Options**

**Volcanic Rocks and VR Vouchers**
- Information
    - VR Vouchers were glorified options for volcanic rocks
    - Time until expiry was 8 - (round #) in in-game days
- Strategy
    - Hedging strategies did not do well; z-score strategies did
    - Fitting a quadratic fit to moneyness vs volatility runs the risk of it being wrong on actual data
    - Create a short rolling window of the IVs on mid prices $\implies$ now, let the mean of this rolling window be used in the z-score computation
    - Delta hedge with underlying after MMing on IV 
        - Predict prices of options with BS model $\implies$ create a bid and ask for each option
        - Market-make like normal EXCEPT keep track of how much you buy / sell of each option, then delta-hedge to minimize delta exposure

**Round 4 - Imports and Exports**

**Macarons**
- Information
    - Macarons can be bought / sold locally or on an island
    - Import tariffs, export tariffs, transport cost, fee for storage (i.e. if possessing a long position)
- Strategy
    - Buy on the island, sell locally (take advantage of the negative import fee)
    - To generate profit, sell_price_local >= buy_price_island + import_fee + transport_fee (find someone who's willing to trade)
